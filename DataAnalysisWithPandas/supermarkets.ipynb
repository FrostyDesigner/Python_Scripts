```python
import pandas
df=pandas.read_csv("supermarkets.csv")
df
```

```python
df["Address"]=df["Address"]+", "+df["City"]+", "+df["State"]+", "+df["Country"]
df
```

```python
df["Coordinates"]=df["Address"].apply(nom.geocode)
df
```

```python
df["latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x != None else None)
df["longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x != None else None)
df
```

```python
import pandas
df=pandas.read_csv("supermarkets.csv")
df
```

```python
#To select rows whose column value equals a scalar, some_value, use ==:
df.loc[df['Address']=="3666 21st St"]
```

```python
#To select rows whose column value is in an iterable, some_values, use isin:
df.loc[df['State']=="CA 94114"]
```

```python
#To select rows whose column value is in an iterable, some_values, use isin:
#df.loc[df['column_name'].isin(some_values)]
df.loc[df['State'].isin(["CA 94114","California 94114"])]
```

```python
#Combine multiple conditions with &:
df.loc[(df['Address'] == "3666 21st St") & df['State'].isin(["CA 94114","California 94114"])]
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A': 'foo bar foo bar foo bar foo foo'.split(),
                   'B': 'one one two three two two one three'.split(),
                   'C': np.arange(8), 'D': np.arange(8) * 2})
```

```python
print(df)
```

```python
print(df.loc[df['B'].isin(['one','three'])])
```

```python
#If you wish to do this many times, it is more efficient to make an index first, and then use df.loc:
df = df.set_index(['B'])
print(df.loc['one'])
```

```python
#to include multiple values from the index use df.index.isin:
df.loc[df.index.isin(['one','two'])]
```
