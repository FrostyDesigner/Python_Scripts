```python
# read the file back as a list of lines
fname="BG1.txt"
fin = open(fname, "r")
data_list = fin.readlines()
fin.close()
#type(data_list)
print(data_list)
```

```python
# now you can access each line (list item) by the index
# the index of a list is zero based
print(data_list[0])  # apple
print(data_list[3])  # melon
```

```python
#loop attempt
for line in data_list:
    print(line)
```

```python
#loop attempt
i=0
cur_dir=None
for line in data_list:
    line=line.replace("\n","")
    #if value is not found in string...
    if line.find("Directory of") == -1:
        print("false")
    else:
        line=line.strip(' \t\n\r')
        cur_dir=line
        print(cur_dir)
    i=i+1
```
